### Using the code from the lesson as a guide and the REST API from https://swapi.dev/ as we did in the lesson, create a dataframe named people that has all of the data for people.

```python
people_url = requests.get('https://swapi.dev/api/people/')
df = pd.DataFrame(people_url.json()['results'])
```

```python
def get_table(name):
    df = pd.DataFrame()
    url = f'https://swapi.dev/api/{name}/'
    while url:
        page = requests.get(url)
        print()
        df = pd.concat([df, pd.DataFrame(page.json()['results'])], axis=0, ignore_index=True)
        url = page.json()['next']
    return df
        
    
```

```python
people = get_table('people')
```

```python
people.head()
```

### Do the same thing, but for planets.

```python
planets = get_table('planets')
```

```python
planets.head()
```

### Extract the data for starships.

```python
starships = get_table('starships')
```

### Save the data in your files to local csv files so that it will be faster to access in the future.

### Combine the data from your three separate dataframes into one large dataframe.

```python
pd.merge(left=people, right=planets, how='left', left_on='homeworld', right_on='url', suffixes=['_people', '_planet'])
```
